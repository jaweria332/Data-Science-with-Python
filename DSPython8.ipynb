```python
import pandas as pd
import numpy as np

#Data series with Pandas
a = pd.Series(['Jiya','Jaweria','Neha','Areeba'], index=['a','b','c','d'])
print(a)
print('\n')
sdata={'Karachi':32,'Islamabad':33,'Lahore':26,'Multan':28}
b=pd.Series(sdata)
print(b)


#DataFrame with Pandas
print('\n')
data={'City':['Karachi','Islamabad','Lahore','Multan','Hyderabad'],'Temperature':[30,31,32,33,31]}
dataset=pd.DataFrame(data)
print(dataset)

#Data frames with row and column label
print('\n')
data=pd.DataFrame((np.arange(16).reshape(4,4)), index=['Lahore', 'Multan', 'Islamabad', 'Karachi'], columns=['one', 'two', 'three', 'four'])
print(data)


#Dropping a row from dataset
print('\n')
data1=data.drop(['Lahore','Multan'])
print(data1)

#Droping a column
print('\n')
data2=data.drop(['one','two'], axis='columns')
print(data2)

#select columns and rows by label
print('\n')
print(data.loc['Islamabad',['one','two']])

#select column and rows by numbers
print('\n')
print(data.iloc[:, :2])
```

# Arithmetic and data allignment 

```python
import pandas as pd
import numpy as np
d1 = pd.Series([2.2,-1.3,8.3,-10.8,7.5],index=['a','b','c','d','e'])
d2 = pd.Series([9.2,5.3,3.8,8.1,-5.5],index=['a','c','d','f','b'])
print(d1 + d2)     #Missing values will propagate in furthur computation

#lets see what happen when we implement it on dataframes
data1=pd.DataFrame((np.arange(16).reshape(4,4)), index=['Lahore', 'Multan', 'Islamabad', 'Karachi'], columns=['one', 'two', 'three', 'four'])
data2=pd.DataFrame((np.arange(16).reshape(4,4)), index=['USA', 'UK', 'Dubai', 'Australia'], columns=['one', 'two', 'three', 'four'])

print(data1)
print(data2)
print(data1 + data2) #since both dataset has different columns so all the values are null u=in their addition

#similarly we can sub two dataframes or matrices
d1 - d2
```

# arithmetic operations with fill values 

```python
a = pd.DataFrame(np.arange(9).reshape(3,3), columns=list('abc'))
b = pd.DataFrame(np.arange(16).reshape(4,4), columns=list('abcd'))

print(a,"\n")
print(b, "\n")
#add element in a 
a.loc[2,'c'] = np.nan

print(a,"\n")
print(a + b)

#some arithmetic operations
print(a.add(b,fill_value=0))

print(a.sub(b))

print(a.mul(b))

print(a.div(b))
```

# Function operation and mapping

```python
#define DataFrame
a = pd.DataFrame(np.random.randn(3,4), columns=list('abcd'), index=['Karachi','Lahore','Islamabad'])

#for absolute values of a
a = np.abs(a)

print(a)

#defining a function
def Max_min_value(x):
    return pd.Series([x.min(), x.max()], index=['Minimum','Maximum'])

a.apply(Max_min_value)  #minimum and maximum from each column
```

# Sorting and Ranking

```python
data = pd.Series(range(4), index=['c','b','a','d'])

#index will sort 
data.sort_index()

dataset = pd.DataFrame(np.arange(12).reshape((3,4)), index=['two','three','one'],columns=[1,2,4,3])
print(dataset)

dataset.sort_index()

print(dataset.sort_index(axis=1)) #sort columns

#to sort a series by its value
dataset2 = pd.Series([1, 9, 3, -1, -8, 10])
print(dataset2.sort_values())

#to sort a Dataframe by its column
print(dataset.sort_values(by=2))

#to sort by multiple columns
dataset.sort_values(by=[1,2,3,4])
```

# Computing descriptive statistics

```python
Data = pd.DataFrame(np.arange(12).reshape((3,4)), index=['one','two','three'], columns=['a','b','c','d'])
print(Data)
print(Data.sum()) #computer column-wise sum
print(Data.idxmax()) #return index of max value in column
print(Data.idxmin()) #return index of min value in column
print(Data.min()) #return min value from each column
print(Data.max()) #return max value from each column
print(Data.describe()) #describe produce multiple summary statistics in one shot
print(Data.count()) #count number of values in column
print(Data.mean()) #compute column-wise mean value
print(Data.prod()) #compute product of all values
print(Data.skew()) #Sample skewness (third moment) of values
print(Data.cumsum()) #compute cumulative sum of values
print(Data.pct_change()) #Compute percent changes
```
